### Site Escolhido foi Sobre o Bruce Lee

##### Importando as Bibliotecas Necessrias 

```python
from requests import get, post, delete #-> Biblioteca requests para acesso a pginas http
from bs4 import BeautifulSoup # -> Para acesso das tags html
```

#### Acessando a pgina http

```python
bruce_resposta_lee = get("https://pt.wikipedia.org/wiki/Bruce_Lee") # -> Usando o mtodo get para acessar a pgina
bruce_resposta_lee # -> Response [200] - Tudo OK.
```

```python
bruce_resposta_lee.text # -> Mtodo texto retorna o contedo do site
```

#### Acessando as Tags HTML

```python
tags = BeautifulSoup(bruce_resposta_lee.text, 'html5lib') # -> Acessando o site e retornando o texto dentro dele no formato html
tags
```

#### Localizar Tag Title

```python
titulo = tags.find('title') # -> Funo que busca a tag passada como parmetro e retorna o seu contedo
titulo # -> Retorno com a Tag HTML title
```

```python
res = []
for tit in titulo.getText().split():
    
    if tit == '':
        break
        
    res.append(tit)    
    
fin = " ".join(res)    
    
print("Pgina Principal: ", fin)
```

#### Localizar Pginas do Wikpedia

```python
lista_url = [] #-> lista para agrupar as urls (href)
lista_nome = [] #-> Lista para agrupar os nomes (titles)

for link in tags.find_all('a'):
    
    lista_url.append(str(link.get('href')))
    lista_nome.append(str(link.get('title')))

print("***********WEB_CRAWLER**********")    
    
for i , j in zip(lista_url, lista_nome): #-> Funo zip para agrupar os itens das duas listas no mesmo loop
    
        if (i.find('wiki') > 0) or (i.find('wikipedia') > 0):
                    
            print("********************************")
            print("PGINA SECUNDRIA:",i, end = "\n")
            print("NONE DA PGINA:", j )
            print("********************************")
```
