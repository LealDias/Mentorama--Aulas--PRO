### Site Escolhido foi Sobre o Bruce Lee

##### Importando as Bibliotecas Necessárias 

```python
from requests import get, post, delete #-> Biblioteca requests para acesso a páginas http
from bs4 import BeautifulSoup # -> Para acesso das tags html
```

#### Acessando a página http

```python
bruce_resposta_lee = get("https://pt.wikipedia.org/wiki/Bruce_Lee") # -> Usando o método get para acessar a página
bruce_resposta_lee # -> Response [200] - Tudo OK.
```

```python
bruce_resposta_lee.text # -> Método texto retorna o conteúdo do site
```

#### Acessando as Tags HTML

```python
tags = BeautifulSoup(bruce_resposta_lee.text, 'html5lib') # -> Acessando o site e retornando o texto dentro dele no formato html
tags
```

#### Localizar Tag Title

```python
titulo = tags.find('title') # -> Função que busca a tag passada como parâmetro e retorna o seu conteúdo
titulo # -> Retorno com a Tag HTML title
```

```python
res = []
for tit in titulo.getText().split():
    
    if tit == '–':
        break
        
    res.append(tit)    
    
fin = " ".join(res)    
    
print("Página Principal: ", fin)
```

#### Localizar Páginas do Wikpedia

```python
lista_url = [] #-> lista para agrupar as urls (href)
lista_nome = [] #-> Lista para agrupar os nomes (titles)

for link in tags.find_all('a'):
    
    lista_url.append(str(link.get('href')))
    lista_nome.append(str(link.get('title')))

print("***********WEB_CRAWLER**********")    
    
for i , j in zip(lista_url, lista_nome): #-> Função zip para agrupar os itens das duas listas no mesmo loop
    
        if (i.find('wiki') > 0) or (i.find('wikipedia') > 0):
                    
            print("********************************")
            print("PÁGINA SECUNDÁRIA:",i, end = "\n")
            print("NONE DA PÁGINA:", j )
            print("********************************")
```
