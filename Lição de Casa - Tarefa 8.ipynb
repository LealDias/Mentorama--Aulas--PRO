#### Importando a biblioteca e as funes que sero utilizadas

```python
from troposphere import Ref, Template, ec2
```

```python
# instnciando a funo template na varivel/objeto template

template = Template()
```

```python
# Criando a instncia

instancia = ec2.Instance(

     "tarefaumserver" , 
    ImageId="ami-00399ec92221878f9",
    InstanceType="t2.micro"

)

instancia_2 = ec2.Instance(

     "tarefadoisserver" , 
    ImageId="ami-00399eb92221878g4",
    InstanceType="t2.micro"

)

instancia_3 = ec2.Instance(

     "tarefatresserver" , 
    ImageId="ami-00699ed90271076f1",
    InstanceType="t2.micro"

)
```

```python
# Adicionando a instncia atravs da funo Template()

template.add_resource(instancia)
template.add_resource(instancia_2)
template.add_resource(instancia_3)
```

#### Gerando o arquivo YAML

```python
# Print na Tela para Verificao

from pprint import pprint

pprint(template.to_yaml())
```

```python
# Gerando o arquivo em formato yaml

with open('tarefa8.yaml' , 'w') as f:
    f.write(template.to_yaml())
```
